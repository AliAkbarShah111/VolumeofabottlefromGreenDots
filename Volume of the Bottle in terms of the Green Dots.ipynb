```python
import cv2
import numpy as np

def main():
    # Load the video
    cap = cv2.VideoCapture(r"C:\Users\Ali\Downloads\greenlight.mp4")

    if not cap.isOpened():
        print("Error opening the video.")
        return

    while True:
        # Capture a frame
        ret, frame = cap.read()

        if not ret:
            print("End of the video.")
            break

        # Convert the frame to HSV
        hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

        # Define the range of green color in HSV
        lower_green = np.array([70, 150, 150])
        upper_green = np.array([85, 255, 255])

        # Threshold the HSV image to get only green colors
        mask = cv2.inRange(hsv, lower_green, upper_green)

        # Bitwise-AND mask and original image
        result = cv2.bitwise_and(frame, frame, mask=mask)

        # Display the original frame and the result
        cv2.imshow('Frame', frame)
        cv2.imshow('Green Detection', result)

        # Break the loop on 'q' key press
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()

if __name__ == "__main__":
    main()
```

```python
import cv2
import numpy as np

def main():
    # Load the video
    cap = cv2.VideoCapture(r"C:\Users\Ali\Downloads\greenlight.mp4")

    if not cap.isOpened():
        print("Error opening the video.")
        return

    while True:
        # Capture a frame
        ret, frame = cap.read()

        if not ret:
            print("End of the video.")
            break

        # Convert the frame to HSV
        hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

        # Define the range of green color in HSV
        lower_green = np.array([35, 100, 100])
        upper_green = np.array([85, 255, 255])

        # Threshold the HSV image to get only green colors
        mask = cv2.inRange(hsv, lower_green, upper_green)

        # Reduce noise
        mask = cv2.erode(mask, None, iterations=2)
        mask = cv2.dilate(mask, None, iterations=2)

        # Find contours in the mask
        contours, _ = cv2.findContours(mask.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

        # Draw detected green dots on the frame
        for contour in contours:
            if cv2.contourArea(contour) < 200:  # Ignore large contours
                continue

            (x, y), radius = cv2.minEnclosingCircle(contour)
            cv2.circle(frame, (int(x), int(y)), int(radius), (0, 255, 0), 2)

        # Display the frame
        cv2.imshow('Green Dots Detection', frame)

        # Break the loop on 'q' key press
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()

if __name__ == "__main__":
    main()
```

```python
import cv2
import numpy as np

def main():
    # Load the video
    cap = cv2.VideoCapture(r"C:\Users\Ali\Downloads\greenlight.mp4")

    if not cap.isOpened():
        print("Error opening the video.")
        return

    while True:
        # Capture a frame
        ret, frame = cap.read()

        if not ret:
            print("End of the video.")
            break

        # Convert the frame to HSV
        hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

        # Define the range of green color in HSV
        lower_green = np.array([35, 100, 100])
        upper_green = np.array([85, 255, 255])

        # Threshold the HSV image to get only green colors
        mask = cv2.inRange(hsv, lower_green, upper_green)

        # Reduce noise
        mask = cv2.erode(mask, None, iterations=2)
        mask = cv2.dilate(mask, None, iterations=2)

        # Find contours in the mask
        contours, _ = cv2.findContours(mask.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

        for contour in contours:
            # Ignore small contours
            if cv2.contourArea(contour) < 200:
                continue

            # Create a mask for this contour
            c_mask = np.zeros_like(mask)
            cv2.drawContours(c_mask, [contour], -1, 255, -1)

            # Calculate the average intensity within this contour
            c_intensity = cv2.mean(frame, mask=c_mask)
            print(f"Average intensity of the green dot: {c_intensity}")

        # Display the frame
        cv2.imshow('Green Dots Detection', frame)

        # Break the loop on 'q' key press
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()

if __name__ == "__main__":
    main()
```
